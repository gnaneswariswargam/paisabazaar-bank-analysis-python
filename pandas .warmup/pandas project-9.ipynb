20. Plot Continuous Variables

```python
cate_value=[]
count_value=[]
for column in df.columns:
    if df[column].nunique() <=10:
        cate_value.append(column)
    else:count_value.append(column)
```

```python
cate_value# categorial values
```

```python
count_value # continous values
```
