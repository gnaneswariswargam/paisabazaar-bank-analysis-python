17.Display Name of The Person Having The Highest BasePay

```python
# Remove leading/trailing spaces
data['BasePay'] = data['BasePay'].astype(str).str.strip()

# Remove commas if any
data['BasePay'] = data['BasePay'].str.replace(',', '', regex=True)

# Convert to numeric, invalid values become NaN
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')
```

```python
print(data['BasePay'].dtype)  # should be float64
print(data['BasePay'].head())
```

```python
max_basepay = data['BasePay'].max()
data[data['BasePay'] == max_basepay]['EmployeeName']
```

18.Find Average BasePay of All Employee Per Year 

```python
# Clean BasePay column
data['BasePay'] = data['BasePay'].astype(str).str.replace(',', '', regex=True)
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')
```

```python
avg_basepay_per_year = data.groupby('Year')['BasePay'].mean()
print(avg_basepay_per_year)
```

19.Find Average BasePay of All Employee Per JobTitle 

```python
# Clean BasePay column (remove commas, spaces) and convert to numeric
data['BasePay'] = data['BasePay'].astype(str).str.replace(',', '', regex=True)
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')
```

```python
avg_basepay_per_job = data.groupby('JobTitle')['BasePay'].mean()
print(avg_basepay_per_job)
```

20.Find Average BasePay of Employee Having Job Title ACCOUNTANT 

```python
# Clean BasePay column
data['BasePay'] = data['BasePay'].astype(str).str.replace(',', '', regex=True)
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')
```

```python
avg_accountant_basepay = data[data['JobTitle'] == 'ACCOUNTANT']['BasePay'].mean()
print("Average BasePay for ACCOUNTANT:", round(avg_accountant_basepay, 2))
```
