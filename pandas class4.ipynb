ADD/REMOVE COLUMNS FROM DATA FRAME


```python
import pandas as pd
d ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,82],
               }
df=pd.DataFrame(d)
df
```

```python
df.drop('Grades', axis=1, inplace=True, errors='ignore')
df
```

6.ASSIGN 

```python
import pandas as pd
d ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,82],
               }
df=pd.DataFrame(d)
df
```
