ADD/REMOVE ROW IN DATA FRAME

```python
import pandas as pd
dict ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
        'Marks':[98,89,99,87,90,83,82],}
df=pd.DataFrame(dict)
df
```

USING CONCAT METHOD : FOR ADDING NEW ROW IN DATA FRAME

```python
df1=pd.DataFrame({'Name':'Anil','Marks':[99]})
df2=pd.concat([df,df1])
df2
```

IF WE ADD TO ANILL IN THE TOP:REVERSE THE SEQUENCE

```python
df2=pd.concat([df1,df])
df2
```

HERE WE HAVE REPATED THE INDEX [0] '2' TIMES SO WE IGNORE THE INDEX

```python
df2=pd.concat([df1,df],ignore_index=True)
df2
```

2.APPEND:USING APPEND METHOD TO ADD THE NEW ROW IN DATA FRAME

```python
import pandas as pd
dict ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
        'Marks':[98,89,99,87,90,83,82],}
df=pd.DataFrame(dict)
df
```

ADD NEW ROW:RAHUL

```python
r={'Name':'Rahul','Marks':89}
df=df.append(r,ignore_index=True)
df
```

APPEND DOESNOT WORK IN NEW VERSION OF PANDAS SO WE USE COCNCAT,LOC METHOD IN ADD IN THE ROW IN DATA FRAME

3.ILOC:USING ILOC WE CAN ADD NEW ROW AT ANY GIVEN POSITION 

```python
import pandas as pd
dict ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
        'Marks':[98,89,99,87,90,83,82],}
df=pd.DataFrame(dict)
df
```

```python
df.iloc[1]=['Anil',89]
df
```

4.LOC:I WANT ADD NEW ROW AT ANY POSITION I.E IS [2]

```python
df.loc[2]=['Rahul',90]
df
```

DROPS THE ROWS

```python
import pandas as pd
dict ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
        'Marks':[98,89,99,87,90,83,82],}
df=pd.DataFrame(dict)
df
```

I WANT TO DROP INDEX=0,1

```python
df.drop([0,1],inplace=True)
df
```

NOW I WANT TO START INDEX WITH 0,1---

```python
df.reset_index(drop=True)
```

REMOVE ROWS [CONDITION]

```python
import pandas as pd
dict ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
        'Marks':[98,89,99,87,90,83,82],}
df=pd.DataFrame(dict)
df
```

I WANT TO REMOVE THE MARKS >90


```python
df.drop(df[df['Marks']>90].index)
```

NOW I WANT TO START INDEX WITH 0

```python
df.reset_index(drop=True)
```
